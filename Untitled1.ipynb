```python
#by default it is acending but if u want descending ascending=False
sortyear=data.sort_values("Year")
sortyear
```

```python
#handling Duplicates
cleanedata=sortyear.drop_duplicates(subset=['Year','Month','State','County'],keep='first',inplace=False)
cleanedata
#.............................................Data Prepration Done...................................
```

```python
#ExploRatory Data Analysis  
```

```python
#Q) Sum of Unemployment in each State and Highest Number of Unemployments In a State in recent years?
# total unemployment rate in each State of America sum tells us the total unemployment that dataet shows
cleanedata.groupby('State')[['Rate']].sum().sort_values('Rate',ascending=False)
```

```python
Q) What is the average rate of unemployment in each State?

# total unemployment rate in each State of America sum tells us the total unemployment that dataet shows mean shows teh avg unemployment rate in each state every year
cleanedata.groupby('State')[['Rate']].mean().sort_values('Rate',ascending=False)
#year wise mean of total unemployment in each year  since 1990 from the most unemployment year till the least
```

```python
# Q.1)What is the rate of change in unemployment each year?
#scatter plot for both the means yearwse and statewise 
sns.set(style="darkgrid")
sns.scatterplot(data=cleanedata,x='Year',y='Rate',s=100).set_title("YearWise Unemployment Rate")
```

```python
Q) What is rate of change of unemployment statewise?
#scatter plot for both the means yearwse and statewise 
sns.set(style="darkgrid")
#sns.countplot(data=cleanedata,y='State').set_title("StateWise Unemployment Rate")
sns.scatterplot(data=cleanedata,x='Rate',y='State',s=100).set_title("StateWise Unemployment Rate")
```
